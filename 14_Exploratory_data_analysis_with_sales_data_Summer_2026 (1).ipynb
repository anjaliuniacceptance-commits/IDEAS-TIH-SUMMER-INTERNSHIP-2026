### Question 1: Import Data and Store-wise Revenue Analysis (4 Marks)

Import `pandas` and `numpy`. Download the dataset `sales.csv` from https://drive.google.com/drive/folders/1gieHICVDBbUKMZiSF4YRQMAJic-w50JM?usp=sharing. Load the dataset `sales.csv` into a DataFrame. Then, find the top 5 stores with the highest total sales and the bottom 5 stores with the lowest total sales.

**Tasks:**
- Aggregate total sales by store_id
- Rank stores by total revenue
- Compare average sales per department within those stores

```python
# Write your answer here
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()

# Load dataset
sales = pd.read_csv('sales.csv')

# Total sales by store
store_sales = sales.groupby('store_id')['weekly_sales'].sum().sort_values(ascending=False)

# Top 5 and Bottom 5 stores
top5_stores = store_sales.head(5)
bottom5_stores = store_sales.tail(5)

print("Top 5 Stores by Revenue")
print(top5_stores)

print("\nBottom 5 Stores by Revenue")
print(bottom5_stores)

# Average sales per department for top stores
top_store_dept_avg = (
    sales[sales['store_id'].isin(top5_stores.index)]
    .groupby(['store_id','department'])['weekly_sales']
    .mean()
)

print("\nAverage Sales per Department (Top Stores)")
print(top_store_dept_avg)

# Average sales per department for bottom stores
bottom_store_dept_avg = (
    sales[sales['store_id'].isin(bottom5_stores.index)]
    .groupby(['store_id','department'])['weekly_sales']
    .mean()
)

print("\nAverage Sales per Department (Bottom Stores)")
print(bottom_store_dept_avg)
```

### Question 2: Department Performance Consistency (3 Marks)

Which departments have the most stable weekly sales and which are the most volatile?

**Tasks:**
- Use standard deviation and coefficient of variation
- Identify highly stable and highly volatile departments

```python
# Write your answer here
# Department-wise sales statistics
dept_stats = sales.groupby('department')['weekly_sales'].agg(['mean', 'std'])

# Coefficient of Variation (CV)
dept_stats['cv'] = dept_stats['std'] / dept_stats['mean']

# Most stable departments (lowest CV)
stable_departments = dept_stats.sort_values(by='cv').head(5)

# Most volatile departments (highest CV)
volatile_departments = dept_stats.sort_values(by='cv', ascending=False).head(5)

print("Most Stable Departments:")
print(stable_departments)

print("\nMost Volatile Departments:")
print(volatile_departments)
```

### Question 3: Holiday vs Non-Holiday Sales (3 Marks)

Analyze whether holidays significantly affect weekly sales.

**Tasks:**
- Compare average sales during holidays vs non-holidays
- Calculate percentage increase/decrease
- Identify departments benefiting most from holidays

```python
# Write your answer here
# Compare average sales during holidays and non-holidays
holiday_sales = sales.groupby('is_holiday')['weekly_sales'].mean()

print("Average Weekly Sales")
print(holiday_sales)

# 0 = Non-Holiday, 1 = Holiday
non_holiday_avg = holiday_sales[0]
holiday_avg = holiday_sales[1]

percent_change = ((holiday_avg - non_holiday_avg) / non_holiday_avg) * 100

print(f"\nPercentage Change in Sales During Holidays: {percent_change:.2f}%")

# Department-wise holiday impact
dept_holiday = sales.groupby(
    ['department', 'is_holiday']
)['weekly_sales'].mean().unstack()

dept_holiday['holiday_impact_%'] = (
    (dept_holiday[1] - dept_holiday[0])
    / dept_holiday[0]
) * 100

top_departments = dept_holiday.sort_values(
    by='holiday_impact_%',
    ascending=False
).head(10)

print("\nDepartments Benefiting Most From Holidays")
print(top_departments)
```

### Question 4: Seasonal Trend Detection (3 Marks)

Identify whether sales increase or decrease during specific months.

**Tasks:**
- Extract month from the date column
- Compute average monthly sales
- Determine highest and lowest sales months

```python
# Write your answer here
# Convert date column to datetime format
sales['date'] = pd.to_datetime(sales['date'])

# Extract month
sales['month'] = sales['date'].dt.month

# Compute average monthly sales
monthly_sales = sales.groupby('month')['weekly_sales'].mean()

print("Average Monthly Sales")
print(monthly_sales)

# Highest and lowest sales months
highest_month = monthly_sales.idxmax()
lowest_month = monthly_sales.idxmin()

print(f"\nHighest Sales Month: {highest_month}")
print(f"Highest Average Sales: {monthly_sales.max():.2f}")

print(f"\nLowest Sales Month: {lowest_month}")
print(f"Lowest Average Sales: {monthly_sales.min():.2f}")
```

### Question 5: Store and Department Contribution (4 Marks)

Which combinations of store and department contribute the most to overall revenue?

**Tasks:**
- Group using store_id and department
- Find top and worst-performing combinations
- Compute percentage contribution

```python
# Write your answer here
# Total sales by store and department
store_dept_sales = sales.groupby(
    ['store_id', 'department']
)['weekly_sales'].sum().reset_index()

# Total revenue across all store-department combinations
total_revenue = store_dept_sales['weekly_sales'].sum()

# Percentage contribution
store_dept_sales['contribution_percent'] = (
    store_dept_sales['weekly_sales'] / total_revenue
) * 100

# Top 10 performing combinations
top_combinations = store_dept_sales.sort_values(
    by='weekly_sales',
    ascending=False
).head(10)

# Bottom 10 performing combinations
worst_combinations = store_dept_sales.sort_values(
    by='weekly_sales'
).head(10)

print("Top Store-Department Combinations")
print(top_combinations)

print("\nWorst Store-Department Combinations")
print(worst_combinations)

print(top_combinations[['store_id',
                        'department',
                        'weekly_sales',
                        'contribution_percent']])

print(worst_combinations[['store_id',
                          'department',
                          'weekly_sales',
                          'contribution_percent']])
```

### Question 6: Load MNIST Data (3 Marks)

Load the MNIST dataset using `sklearn.datasets.load_digits`. Separate the dataset into features (`X`) and target labels (`y`).
Print the shape of the features and the target arrays.

**Expected Output:** The shape of `X` and `y`.

```python
# Write your answer here
from sklearn.datasets import load_digits

# Load MNIST-like digits dataset
digits = load_digits()

# Separate features and target
X = digits.data
y = digits.target

# Print shapes
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

### Question 7: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
# Write your answer here
from sklearn.cluster import KMeans

# Initialize K-Means
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)

# Fit the model
kmeans.fit(X)

# Get cluster labels
kmeans_labels = kmeans.labels_

print("K-Means model fitted successfully!")
print("Cluster Labels:")
print(kmeans_labels)
```

### Question 8: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
# Write your answer here
from scipy.stats import mode
from sklearn.metrics import f1_score
import numpy as np

# Create array for mapped labels
mapped_labels = np.zeros_like(kmeans_labels)

# Map each cluster to the most common true digit label
for cluster in range(10):
    mask = (kmeans_labels == cluster)

    if np.sum(mask) > 0:
        most_common_label = mode(y[mask], keepdims=True).mode[0]
        mapped_labels[mask] = most_common_label

# Calculate Macro F1 Score
f1 = f1_score(y, mapped_labels, average='macro')

print("Macro F1 Score:", f1)
```
